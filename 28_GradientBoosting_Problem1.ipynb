```python
import pandas as pd
from sklearn.ensemble import GradientBoostingRegressor

def train_gradient_boosting_regressor(X, y, n_estimators=2, learning_rate=0.5, max_depth=1, random_state=42):
    model = GradientBoostingRegressor(
        n_estimators=n_estimators,
        learning_rate=learning_rate,
        max_depth=max_depth,
        random_state=random_state
    )
    model.fit(X, y)
    return model

def main():
    # Example dataset
    data = {
        'Age': [22, 25, 28, 35, 45],
        'Experience': [1, 2, 4, 6, 10],
        'Income': [30, 35, 45, 60, 80]
    }
    df = pd.DataFrame(data)
    X = df[['Age', 'Experience']]
    y = df['Income']

    # Train model
    gbr = train_gradient_boosting_regressor(X, y)

    # Predict for new data
    new_data = pd.DataFrame({'Age': [40], 'Experience': [8]})
    pred = gbr.predict(new_data)
    print(f"Predicted Income for Age=40, Experience=8: {pred[0]:.2f}")

    # Optionally, predict for multiple new samples
    # new_samples = pd.DataFrame({'Age': [30, 50], 'Experience': [3, 12]})
    # preds = gbr.predict(new_samples)
    # print("Predictions for new samples:", preds)

if __name__ == "__main__":
    main()
```
